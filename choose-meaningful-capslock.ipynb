```python
import pandas as pd
from string import punctuation as punct
import os
```

```python
files = os.listdir("./WikiExtractor/")
files = [i for i in files if not i.startswith(".") and not i.endswith(".py") and i not in ["wiki-files", "wiki.txt"]]
print(files)
```

```python
len(files)
```

```python
df = pd.DataFrame()

for i in files:
    df1 = pd.read_csv("./WikiExtractor/{}".format(i), 
                 sep="\t", 
                 names=["word", "metrics", "count"])

    words = df1.word.values.tolist()
    df1["word"] = [i.strip(punct).replace("Ё", "Е") for i in words]
    df = df.append(df1)
```

```python
df["upper_count"] = round(df["metrics"] * df["count"], 2)
```

```python
df.head(20)
```

```python
df = df.groupby(by="word").sum().reset_index()
```

```python
df["metrics"] = round(df["upper_count"]/df["count"],2)
```

```python
usual_capslock = df[(df.metrics >= 0.9) & (~df.metrics.isna())].word.values.tolist()
```

```python
len(usual_capslock)
```

```python
unusual_capslock = df[df.metrics < 0.9].word.values.tolist() # осмысленный капслок
```

```python
len(unusual_capslock)
```

```python
unusual_capslock
```

```python
rest = df[(~df.word.isin(usual_capslock)) & (~df.word.isin(unusual_capslock))].word.values.tolist()
```

```python
with open("rest_caps.txt", "r") as f:
    f = f.readlines()
    f = [i.strip() for i in f]
print(len(f))
```

```python
unusual_capslock += f
print(len(unusual_capslock))
```
